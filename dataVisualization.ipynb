```python
data = pd.read_csv('ratings_small.csv')
data
```

```python
data.columns
```

creating a seprate dataset for the avgRating

```python
avg_rating_movie_data=data[['movieId', 'rating']]
avg_rating_movie_data
```

```python
data.hist(bins=60)
```

Generating line graph between the movieId and rating

```python
movieMoreThan4rating=data[data['rating']>4]#here I selected rating above 4

movieMoreThan4rating[['movieId','rating']].plot(x='movieId',y='rating',figsize=(20,10),kind='line')
```

```python
import matplotlib.pyplot as plt

rating=avg_rating_movie_data['rating']
plt.hist(rating)
plt.xlabel('Rating')
plt.ylabel('Number of Movies which got this rating')
plt.title('Distribution of Movie Ratings')

# now for my own clearification I am writing this down as you can see there are multiple dips between 2 points in x axis this
# is beacuse the rating is in float not int there are ratings like 4.5 , 1.5 etc
```

```python
data.plot(kind='scatter', x='timestamp', y='rating')
plt.xlabel('timestamp')
plt.ylabel('rating')
plt.title('Movie Ratings VS timestamp')
```

```python
data['movieId'].plot(kind='line', figsize=(8, 4), title='movieId')

plt.xlabel('movieId')
plt.ylabel('Number of movieId in that range')
plt.title('Movie Ratings description')
```
